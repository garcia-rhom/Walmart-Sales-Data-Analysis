# 3. Data Cleaning

## **a. Inconsistent column name formatting**

```python
print(df.columns)
```

```python
df.columns = df.columns.str.lower()
print(df.columns)
```

## **b. Incorrect data type for date and time columns**

```python
df['datetime'] = pd.to_datetime(df['date'] + ' ' + df['time'].astype(str))
df.drop(columns=['time', 'date'], inplace=True)
df = df[[*df.columns[:6], 'datetime', *df.columns[6:-1]]]
df.info()
```

## **c. 31 null values in the unit_price and quantity columns.**

```python
overall_null_percentage = df.isnull().sum().sum() / df.size * 100
print(f"Overall null percentage: {overall_null_percentage:.2f}%")
```

```python
df = df.dropna()
df.isnull().sum()
```
